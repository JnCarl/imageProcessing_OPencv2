```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread('./Coldplayx&y.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar el umbral binario
umbral, imagen_umbral = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY)

# Mostrar la imagen original y la imagen con umbral
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Imagen con Umbral Binario')
plt.imshow(imagen_umbral, cmap='gray')

plt.show()
```
